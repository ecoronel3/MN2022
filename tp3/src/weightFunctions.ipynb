```python
import numpy as np
import math
import numpy.linalg as linalg
import matplotlib.pyplot as plt

def tricubic(x):
    y = np.zeros_like(x)
    idx = (x >= -1) & (x <= 1)
    y[idx] = np.power(1.0 - np.power(np.abs(x[idx]), 3), 3)
    return y

def bisquare(x):
    y = np.zeros_like(x)
    idx = (x >= -1) & (x <= 1)
    y[idx] = np.power(1.0 - np.power(x[idx], 2), 2)
    return y
    
def quintuple(x):
    y = np.zeros_like(x)
    idx = (x >= -1) & (x <= 1)
    y[idx] = np.power(1.0 - np.power(np.abs(x[idx]), 5), 5)
    return y
```

```python
X = np.linspace(-1, 1, 200)

fig = plt.figure(figsize=(10, 5))
ax1 = fig.add_subplot(111)

ax1.plot(X, bisquare(X), label='bisquare')
ax1.plot(X, tricubic(X), label='tricubic')
ax1.plot(X, quintuple(X), label='quintuple')
ax1.legend(loc='lower left')
ax1.grid()


```
